```python
def check_input(element):
    return isinstance(element, str)


def get_substring(s):
    seen = set()
    left = 0
    max_len = 0
    start = 0

    for right in range(len(s)):
        while s[right] in seen:
            seen.remove(s[left])
            left += 1
        seen.add(s[right])
        if right - left + 1 > max_len:
            max_len = right - left + 1
            start = left

    return s[start:start + max_len]


def main(arr):
    max_length = -1
    index = -1
    longest_string = ""

    for i, elem in enumerate(arr):
        if check_input(elem):
            if len(elem) > max_length:
                max_length = len(elem)
                index = i
                longest_string = elem

    if index == -1:
        print("-1, ''")
        return

    substring = get_substring(longest_string)
    print(f"{index}, '{substring}'")
```
